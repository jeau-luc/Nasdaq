##Faire le graphique de l'action XOM sur la période 1980 à 2020

```python
plt.figure(figsize=(16,8))
plt.plot(data.date,data.close)
plt.xlabel('Dates')
plt.ylabel('Clotures (en $)')
plt.title("Evolution de l'action XOM sur la période 1980 à 2020")

;
```

```python
plt.figure(figsize=(10,8))
plt.hist(data.close, color='green')
plt.xlabel('Nombre')
plt.ylabel('Clotures (en $)')
plt.title("Histogramme de l'action XOM sur la période 1980 à 2020")
```

```python
plt.figure(figsize=(10,8))
plt.hist(data.close, color='red',orientation='horizontal',align='right')
plt.ylabel('Nombre')
plt.xlabel('Clotures (en $)')
plt.title("Histogramme de l'action XOM sur la période 1980 à 2020")
```

```python
plt.figure(figsize=(10,8))
plt.hist2d(data.open,data.close, bins=10)
plt.colorbar()
plt.title("Histogramme de l'action XOM sur la période 1980 à 2020")
```

```python
from matplotlib.colors import LogNorm
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(14,8))
# normal distribution center at x=0 and y=5
x = np.random.randn(100000)
y = np.random.randn(100000) + 5

plt.hist2d(x, y, bins=40, norm=LogNorm())
plt.colorbar()
plt.show()
```

```python
from matplotlib import colors 
import numpy as np 
from numpy.random import multivariate_normal 
import matplotlib.pyplot as plt 

plt.figure(figsize=(14,8)) 
    
result = np.vstack([ 
    multivariate_normal([10, 10], 
            [[3, 2], [2, 3]], size=1000000), 
    multivariate_normal([30, 20], 
            [[2, 3], [1, 3]], size=100000) 
]) 
  
plt.hist2d(result[:, 0], 
           result[:, 1], 
           bins = 100,  
           cmap = "Greens", 
           norm = colors.LogNorm()) 
plt.title('matplotlib.pyplot.hist2d function / Example') 
plt.show() 

plt.figure(figsize=(14,8))

plt.hist2d(result[:, 0],  
           result[:, 1], 
           bins = 100,  
           cmap = "RdYlGn_r", 
           norm = colors.LogNorm()) 
plt.show() 
```
